```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv("results.csv")
```

```python
# Map the numeric 'Case' values to descriptive strings
case_mapping = {
    1: 'No ST',
    2: 'No Zero ST',
    3: 'Predicted ST',
    4: 'Tuned Predicted ST'
}

df['Case'] = df['Case'].map(case_mapping)
```

```python
df
```

```python
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))  # 2 rows, 2 cols
axes = axes.flatten()  # Make it easier to index

for i, metric in enumerate(metrics):
    sns.boxplot(x="Case", y=metric, data=df, ax=axes[i])
    axes[i].set_title(f"{metric} Distribution by Case", fontsize=12)
    axes[i].tick_params(axis='x', rotation=15)  # Tilt x-axis labels
    axes[i].set_xlabel("")  # Optionally remove x-axis label
    axes[i].set_ylabel(metric)

plt.tight_layout()
plt.show()
```
